# Simple Spectra Tools

```python
%matplotlib notebook
```

```python
# imports
import numpy as np
from matplotlib import pyplot as plt

from linetools.spectra import xspectrum1d 
```

# Dummy spectrum

```python
wave = 5000. + np.arange(3000)
wave
```

```python
flux = np.random.uniform(size=wave.size)
flux
```

# Using matplotlib (boring)

```python
plt.clf()
ax = plt.gca()
ax.plot(wave, flux)
#
ax.set_xlabel('Wavelength (Ang)')
ax.set_ylabel('Flux')
#
plt.show()
```

# Using linetools
    https://linetools.readthedocs.io/en/latest/

```python
xspec = xspectrum1d.XSpectrum1D.from_tuple((wave,flux))
```

## In the Notebook

```python
xspec.plot()
```

## The real deal
    See these docs:  https://github.com/linetools/linetools/blob/master/docs/xspecgui.rst

```python
xspec.plot(xspec=True)
```
